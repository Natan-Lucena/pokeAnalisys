```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("pokemon.csv")
```

```python
data.head()
```

# Project's goal:
### This project seeks to analyze a database about pokemons up to generation 6

```python
dataGen1 = np.mean (data[data["Generation"]== 1 ]["Total"])
dataGen2 = np.mean (data[data["Generation"]== 2 ]["Total"])
dataGen3 = np.mean (data[data["Generation"]== 3 ]["Total"])
dataGen4 = np.mean (data[data["Generation"]== 4 ]["Total"])
dataGen5 = np.mean (data[data["Generation"]== 5 ]["Total"])
dataGen6 = np.mean (data[data["Generation"]== 6 ]["Total"])
info = [dataGen1,dataGen2,dataGen3,dataGen4,dataGen5,dataGen6]
gensDatas = [dataGen1,dataGen2,dataGen3,dataGen4,dataGen5,dataGen6, np.mean(info) ]
genNames = [1,2,3,4,5,6, "average"]
```

## First analysis:
The evolution of pokemon stats over generations

```python
plt.xlabel("Gen")
plt.ylabel("Status")
plt.title("Status average comparison")

plt.plot(genNames,gensDatas, color = "Red",linewidth = 5,linestyle = "dotted" )

```

## Chart Information:
The graph shows that there is a considerable variation between one generation and another, since from the first generation to the second there is a considerable drop in the average status of pokemons. However, after the second generation, the value rises to generation 4 (being the generation with the highest average Status). After the peak of generation 4, there is a considerable drop in generation 5, which is more than 20 points behind generation 4 (but even greater than generation 1 and 2). Generation 6, on the other hand, was the one that presented the smallest variation when compared to its previous generation, having risen a little more than 1 point

## Analysis:
It can be seen that there has been a large increase in the value of pokemons' status since generation 1, only generation 2 being behind the original. In addition, it is clear that generation 4 was where the greatest value is found. In addition, it is worth mentioning that from generation 5 to 6 was where the smallest variation from one generation to another occurred. When compared with the general average of the values, generations 1, 2 and 5 are below expectations.

```python
dataLends = data[data["Legendary"] == True]["Total"]
dataNoLend = data[data["Legendary"] == False]["Total"]
dataTotal = data.loc[:,"Total"]
infosLends = [np.mean(dataLends),np.mean(dataNoLend), np.mean(dataTotal)]
namesLends = ["Legendary", "NoLegendary" , "AllPokemon"]
colorsLend = ["Red", "Blue" , "Gray"]
```

## Second analisys: 
comparing legendary pokemons with non-legendary pokemons

```python
fig, ax = plt.subplots()
ax.bar(namesLends,infosLends , color = colorsLend)
ax.set_ylabel("Total of Stats")
ax.set_title("Legendary Pokemon and Non-Legendary Pokemon")
```

## Chart information
The graph presents 3 pieces of information, the average of the total Status of the legendary Pokemons, the non-legendary ones and all the Pokemons, from that, some analyzes can be carried out.

## Analysis:
It can be observed that the average of the legendary pokemons is much higher than that of the non-legendary pokemons, approximately 200 points, from this, it can be inferred that they are more powerful in the franchise games. Another important information that can be taken from the graph is that when analyzing only the non-legendary pokemons, they are behind the general average (which also includes the legendary ones), from this, it is clear that the legendary ones were important to make the overall average go up.
Another important perception is that when compared with the general average, the legendary pokemons are still way ahead, approximately 200 points, however, the difference is smaller than when they are compared only with the non-legendary pokemons.

```python

dataTypes = []
type_names = ["Fire", "Bug", "Dark", "Dragon", "Electric", "Fairy", "Fighting", "Flying", "Ghost", "Grass", "Ground", "Ice", "Normal", "Poison", "Psychic", "Rock", "Steel", "Water"]
cores = ['#FF5733', '#9ACD32', '#2F4F4F', '#9400D3', '#FFFF00', '#FF69B4', '#8B0000', '#87CEFA', '#483D8B', '#008000', '#D2B48C', '#00FFFF', '#A9A9A9', '#8A2BE2', '#FF1493', '#A0522D', '#808080', '#00BFFF']
for tipo in type_names:
    dataInfo = np.mean(data[(data["Type 1"] == tipo) | (data["Type 2"] == tipo)]["Total"])
    dataTypes.append(dataInfo)
sortedDataType = sorted(dataTypes)
arrayTipada = []
niceColors = []

for indexado in sortedDataType:
    index = dataTypes.index(indexado)
    arrayTipada.append(type_names[index])
    niceColors.append(cores[index])
```

```python
fig, ax = plt.subplots()
ax.bar(arrayTipada, sortedDataType, color = niceColors)
plt.xticks(rotation=70)
plt.title("Avarege of stats")
ax.set_ylabel("Avarage")
```
